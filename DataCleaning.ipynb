# DATA CLEANING

```python
# 1. REMOVING DUPLICATES
df = df.drop_duplicates()
```

```python
# CHECKING, DUPLICATES ARE REMOVED OR NOT
df.duplicated().sum()
```

```python
# 2. Standarizing Data
```

```python
# REMOVING WHITE SPACES
df.loc[:, "First_Name"] = df["First_Name"].str.strip("")
```

```python
# CHECKING, WHITE SPACES ARE REMOVED OR NOT
df["First_Name"]
```

```python
# REMOVING THE UN-NECESSARY CHARACTERS
df.loc[:, "Last_Name"] = df["Last_Name"].str.strip("/._")
```

```python
# CHECKING THE DESIRED CHARACTERS ARE REMOVED OR NOT
df["Last_Name"]
```

```python
# REPLACING OLD VALUE WITH NEW VALUE
df.loc[:, "Phone_Number"] = df["Phone_Number"].str.replace(r"[^\d]","", regex = True)
```

```python
# CONVERTING OBJECT VALUES INTO STRING VALUES USING LAMBDA
df.loc[:, "Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
```

```python
# ADDING "-" AFTER EVERY THREE DIGITS
df.loc[:, "Phone_Number"] = df["Phone_Number"].apply(lambda x : x[0:3] + "-" + x[3:6] + "-" + x[6:10])
```

```python
# REPLACING OLD VALUE WITH NEW VALUE
df.loc[:, "Phone_Number"] = df["Phone_Number"].str.replace("nan--", "")
```

```python
# REPLACING OLD VALUE WITH NEW VALUE
df.loc[:, "Phone_Number"] = df["Phone_Number"].str.replace("--", "")
```

```python
# SPLITTING THE VALUES OF ONE COLUMN INTO IN THREE COLUMNS
df.loc[:, ["Street_Address","State","Zip_Code"]] = df["Address"].str.split(",", n = 2, expand = True)
```

```python
# REPLACING OLD VALUE WITH NEW VALUE
df.loc[:, "Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
```

```python
# REPLACING OLD VALUE WITH NEW VALUE
df.loc[:, "Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
```

```python
# REPLACING OLD VALUE WITH NEW VALUE
df.loc[:, "Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
```

```python
# REPLACING OLD VALUE WITH NEW VALUE
df.loc[:, "Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
```

```python
# DROPPING ROWS IN COLUMN EQUAL TO "Y"
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == "Y":
        df.drop(x, inplace = True)
```

```python
# DROPPING ROWS IN COLUMN HAVING WHITE SPACES
for x in df.index:
    if df.loc[x, "Phone_Number"] == "" :
        df.drop(x, inplace = True)
```

```python
# 3. HANDLING NULL/MISISNG VALUES
```

```python
# REPLACING N/a WITH WHITE SPACES
df = df.replace("N/a", "")
```

```python
# FILLING NULL VALUES WITH WHITE SPACES
df = df.fillna("")
```

```python
# DROPPING ROWS IN COLUMN HAVING WHITE SPACES
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == "" :
        df.drop(x, inplace = True)
```

```python
# RESETTING INDEX
df = df.reset_index(drop = True)
```

```python
# REMOVING UN-NECESSARY COLUMNS
df.drop(columns = ["Address", "Not_Useful_Column", "Unnamed: 8", "Unnamed: 9"], inplace = True)
```
